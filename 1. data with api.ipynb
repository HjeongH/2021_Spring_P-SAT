# DATA wiht API
### 공공데이터포털 api 이용 신청

```python
#공공데이터 미리보기의 url과 설명문서를 참고하면 쉽게 찾을 수 있다.
url = 'apirl주소?serviceKey=본인서비스키&bgnde=시작일&endde=종료일'


request = requests.get(url).text.encode('utf-8')
xmlobj = bs4.BeautifulSoup(request, 'lxml-xml')
rows = xmlobj.findAll('item')
```

```python
#변수명으로 데이터프레임 형태 만들기
nameList = []
columns = list(rows[0])
for j in range(0,22):
    nameList.append(columns[j].name)
    
data = pd.DataFrame(columns = nameList)
data
```

```python
# 데이터프레임에 정보 추가
# 데이터마다 정보량 차이가 있어서 경우를 나눠 저장함

i=0
while i <= 128884:
    infos = list(rows[i])
    details=[]
    
    if len(infos) == 22:
        for j in range(0, 22):
            eachinfo = infos[j].text
            details.append(eachinfo)
            
    if len(infos) == 21:
        for j in range(0, 4):
            eachinfo = infos[j].text
            details.append(eachinfo)
            
        details.append(".")
        
        for j in range(4, 21):
            eachinfo = infos[j].text
            details.append(eachinfo)
            
    if len(infos) == 23:
        for j in range(0, 12):
            eachinfo = infos[j].text
            details.append(eachinfo)
        for j in range(13, 23):
            eachinfo = infos[j].text
            details.append(eachinfo)    
            
            
    data.loc[len(data)] = details
    
    i += 1
    
data.shape
```

```python
#분석에 불필요한 변수 제거
cleaned_data=data.drop(['careTel','chargeNm','filename','happenDt','happenPlace','officetel','popfile'],axis=1)
```

```python
#csv파일로 저장하기
cleaned_data.to_csv('파일명.csv', index=False, header=True,encoding='euc-kr')
```

```python
#개/고양이 품종 나눠서 csv파일 저장하기
data_dog = cleaned_data[cleaned_data['kindCd'].str.contains("개")]
data_cat = cleaned_data[cleaned_data['kindCd'].str.contains("고양이")]
data_dog.to_csv('data_dog.csv', index=False, header=True,encoding='euc-kr')
data_cat.to_csv('data_cat.csv', index=False, header=True,encoding='euc-kr')
```
